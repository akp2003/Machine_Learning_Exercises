# Preprocessing

```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("ParsiAI/snappfood-sentiment-analysis")
```

```python
train = ds['train'].to_pandas()
```

```python
train.head()
```

```python
train["comment"].apply(lambda x : len(x)).describe()
```

```python
print("Does label match with label_id? " + "✅" if (train["label"].apply(lambda x : x == "SAD") == train["label_id"]).prod() else "❌")
```

```python
print("There is no string of just space characters: " + "✅" if train["comment"].apply(lambda x : re.fullmatch(r"\s*",x) == None).prod() else "❌")
```

```python
print("There is no missing data: " + "✅" if not train.isna().prod().prod() else "❌")
```
