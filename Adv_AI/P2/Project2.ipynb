```python
import graphviz

S_start = (0,0)

def produce_grid(m,n, size=5):
    g = graphviz.Digraph(comment='The Grid', engine = 'neato')
    g.graph_attr.update({'overlap' : 'false', 'size' : str(size), 'ratio': 'fill'})
    for i in range(m+1):
        for j in range(n+1):
            g.node(f'({i},{j})', f"({i},{j})", pos=f'{i},{j}!')
    for i in range(m+1):
        for j in range(n+1):
            if j+1 <= n:
                g.edge(f'({i},{j})', f'({i},{j+1})', label=f'{(1+max(0,i))}')
            if i-1 >= 0:
                g.edge(f'({i},{j})', f'({i-1},{j})', label=f'{(1+max(0,i))}')
            if i+1 <= m:
                g.edge(f'({i},{j})', f'({i+1},{j})', label=f'{(1+max(0,i))}')
            if j-1 >= 0:
                g.edge(f'({i},{j})', f'({i},{j-1})', label=f'{(1+max(0,i))}')
    g.node(f'({S_start[0]},{S_start[1]})', color  = 'green', style='filled')
    g.node(f'({m},{n})', color  = 'red', style='filled')
    return g
```

### 0.a

Look at the following example 

```python
produce_grid(m = 2, n = 3)
```

As you can see we can intuitively say that the minimum cost for traveling from `(0,0)` to `(m,n)` is

$$ n + \frac{m(m+1)}{2} $$

```python
def min_cost(m,n):
    return n + m*(m + 1)//2
min_cost(2,3)
```

The path is unique, you always have to go `n` step up and then `m` steps right.
