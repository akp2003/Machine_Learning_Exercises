```python
import graphviz

S_start = (0,0)

def produce_grid(m,n, size=5):
    g = graphviz.Digraph(comment='The Grid', engine = 'neato')
    g.graph_attr.update({'overlap' : 'false', 'size' : str(size), 'ratio': 'fill'})
    for i in range(m+1):
        for j in range(n+1):
            g.node(f'({i},{j})', f"({i},{j})", pos=f'{i},{j}!')
    for i in range(m+1):
        for j in range(n+1):
            if j+1 <= n:
                g.edge(f'({i},{j})', f'({i},{j+1})', label=f'{(1+max(0,i))}')
            if i-1 >= 0:
                g.edge(f'({i},{j})', f'({i-1},{j})', label=f'{(1+max(0,i))}')
            if i+1 <= m:
                g.edge(f'({i},{j})', f'({i+1},{j})', label=f'{(1+max(0,i))}')
            if j-1 >= 0:
                g.edge(f'({i},{j})', f'({i},{j-1})', label=f'{(1+max(0,i))}')
    g.node(f'({S_start[0]},{S_start[1]})', color  = 'green', style='filled')
    g.node(f'({m},{n})', color  = 'red', style='filled')
    return g
```

#### Q1 : Adding edges cannot increase shortest-path distances.

How do you define distance. The statement can be **FALSE**: Consider the following example 


```python
g = graphviz.Digraph(graph_attr={'rankdir':'LR'})
g.node('A') ; g.node('B') ; g.node('C')
g.edge('A','B',label='3') ; g.edge('B','C',label='5')
g.edge('A','C',label='7')
g
```

If you define distance, the number of steps that we have to take to travel with minimum cost then here `shortest-path distances` is 1 and `minimum cost` is 7.

But if we add an edge

```python
g.edge('B','C', label = '1'); g
```

Then the `shortest-path distances` is 2 and `minimum cost` is 4.

As you can see the `shortest-path distances` with this particular definition increased!

**But the `minimum cost` never increases.**

If you define `shortest-path distances` to be equal to `minimum cost` then the statement is **TRUE**.
